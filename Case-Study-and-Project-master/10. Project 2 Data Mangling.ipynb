### 8. Make country as the first column of the dataframe.

```python
#function to rearrange the columns

def change_column_order(df, col_name, index):
    cols = df.columns.tolist()
    cols.remove(col_name)
    cols.insert(index, col_name)
    return df[cols]
```

### 17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
#adding key column to product
sLength = len(df_product['ProductID'])
df_product['key'] = pd.Series(np.ones(sLength)) 

#adding key column to user
sLength = len(df_user['UserID'])
df_user['key'] = pd.Series(np.ones(sLength)) 

#Cross Product
df_cross = pd.merge(df_user, df_product,on='key')[['UserID', 'ProductID']]
df_cross
```
